## Get the data.

```r
library(tidyverse) # Always need to load libraries first
dialogs <- read_csv("https://raw.githubusercontent.com/Vincent-Toups/bios512/fcbc65a2696c7cff80d0f6ed1dd5c97abf0ef800/effective-visualization/source_data/tng.csv") # Reading CSV directly from GitHub
head(dialogs, 10) # Showing first 10 observations
```

```r
names(dialogs)
```

```r
dialogs %>% group_by(character) %>% tally() %>% arrange(desc(n))
```

Let's see who speaks the longest on average.   
First, add a `dialog_length` column by mutating our data set and using the `str_length()` function to get the length of each piece of dialog.  
Then, we can make a table that shows the mean dialog length for each character.  

```r
dialogs %>% mutate(dialog_length=str_length(dialog)) %>% group_by(character) %>% summarize(mean_dialog_length = mean(dialog_length)) %>% arrange(desc(mean_dialog_length))
```

Notice, there are some issues in this data set. We see [CHARACTER], [CHARACTER]'s voice, etc...
Let's troubleshoot.

## Fix weird data.

Let's check on what "BEVERLY'S" refers to... If we look at the source code, this episode says that this refers to Beverly's voice log, and should be treated as normal dialog. 

```r
dialogs %>% filter(character=="BEVERLY'S")
```

Then, using a bunch of mutate statements and regular expressions, we'll change delete the extra stuff that creates new characters.

**Regular Expressions** (that we used):  
Regular expressions in R are powerful tools used for pattern matching and manipulation within strings.  
. = any single character except for a newline character  
\* = quantifier that matches the preceding element zero or more times  
$ = represents the end of a string

In the end, we just filtered to the main characters.

```r
dialogs_fixed <- dialogs %>%
  mutate(
    character = str_replace_all(character, "'S.*$", ""),
    character = str_replace_all(character, " VOICE", ""),
    character = str_replace_all(character, "\\.", ""),
    character = str_replace_all(character, "'", ""),
    character = str_replace_all(character, "S COM", ""),
    character = str_replace_all(character, " COM", ""),
    dialog_length = str_length(dialog)
  ) %>%
  filter(character %in% unlist(str_split("PICARD RIKER DATA TROI BEVERLY WORF WESLEY GEORDI", " ")))
```

## Summarize/show the data.

First, we can look at the mean and standard deviation dialog length per character.

```r
dialogs_fixed %>% group_by(character) %>% summarize(mean_dialog_length = mean(dialog_length), std_dialog_length=sd(dialog_length)) %>% arrange(desc(mean_dialog_length))
```

We can add an additional grouping by episode number to get some more info.

```r
dialog_len_per_ep <- dialogs_fixed %>% group_by(character, episode_number) %>% summarize(mean_dialog_length = mean(dialog_length), std_dialog_length=sd(dialog_length)) %>% arrange(desc(mean_dialog_length))

dialog_len_per_ep
```

## Plot the data!
Let's check on the density of the dialog lengths.

```r
ggplot(dialogs_fixed) + geom_density(aes(x=dialog_length))
```

Next, let's make a boxplot of the dialog length by character. We can arrange these in order by making the character a factor. 

```r
# Factor variable
for_factor <- dialog_len_per_ep %>% group_by(character) %>% summarise(m=mean(mean_dialog_length)) %>% arrange(desc(m))

ggplot(dialog_len_per_ep, aes(factor(character,for_factor$character), mean_dialog_length)) + geom_boxplot()
```

Let's compare count vs. mean dialog length for each character in a scatterplot.

```r
dialog_len_per_ep <- dialogs_fixed %>% 
    group_by(character, episode_number) %>% 
    summarize(mean_dialog_length = mean(dialog_length), dialog_count=n()) %>% 
    arrange(desc(mean_dialog_length))

ggplot(dialog_len_per_ep, aes(dialog_count, mean_dialog_length)) + geom_point(aes(color=character)) + facet_wrap(~character)
```
