# Lecture 20: Intro To Shiny - Examples Only
## Requirements
To run these examples, you'll need the following R packages installed: `tidyverse`, `shiny`, and `plotly`. 
To install them on your terminal, type the following:
1. `R` into your terminal.
   -  It will prompt you to select a CRAN mirror. Type the number corresponding to the mirror, then it enter.
3. `install.packages(c("tidyverse", "shiny", "plotly"))` 
4. `q()` to quit the `R` session.
Additionally, you'll need all the R files in this directory, and CSV files in the derived_data subdirectory. 

## General Format To Run
To run the examples, type the following into the terminal:
1. `Rscript <filename>.R`.
2. Go to http://localhost:PORT/ on your web browser.
3. Type `CTRL+C` in the terminal to exit.

## Examples In Order
#### 1) Hello World (`shiny-hello.R`)
This one can be run in two ways.  
  
Using the shell scripts `start.sh` or `start_silicon.sh`:  
In your terminal, type:
1. `chmod +x start.sh`
2. `./start.sh` 
3. Go to http://localhost:8123/. You should see text that says "Hello World".

If you want to use the R script only, type:
1. `Rscript shiny-hello.R`
2. Go to http://localhost:8123/.
3. Type `CTRL+C` in the terminal to exit.
